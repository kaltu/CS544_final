```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from tqdm.auto import tqdm, trange

reduced_text = []
for i, doc in tqdm(enumerate(note_events['text']), total=len(note_events['text'])):
    if type(doc) != str:
        reduced_text.append('')
        continue
    zen = TextBlob(doc)
    reduced_sentences = []
    # split sentences
    for sentence in zen.sentences:
        if 'sepsis' in sentence or 'septic' in sentence:
            for i, word in enumerate(sentence.words):
                if 'sepsis' in word or 'septic' in word:
                    tgt = ' '.join(sentence.words[i-10:i+12]).strip()
                    reduced_sentences.append(tgt if 'sepsis' in tgt or 'septic' in tgt else '')
    reduced_text.append('\n'.join(reduced_sentences) if reduced_sentences else '')
note_events['reduced_text'] = reduced_text
```

```python
note_events
```

```python
dropped = note_events.drop_duplicates(keep=False, subset=['reduced_text'])
dropped
```

```python
def textblob_tokenizer(text):
    return [word.lemmatize() for word in TextBlob(text).words]
```

```python
vectorizer = TfidfVectorizer(tokenizer=textblob_tokenizer)
X = vectorizer.fit_transform(note_events['reduced_text'])
X
```

```python
# Apply K-Means clustering
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Assign the cluster labels to the documents
doc_clusters = kmeans.predict(X)
doc_clusters
```

```python
note_events['cluster'] = doc_clusters
note_events
```

```python
note_events['cluster'].value_counts()
```

```python
grouped = note_events.groupby('cluster')

for cluster_label, cluster_df in grouped:
    print(f"First 5 rows for cluster {cluster_label}:")
    first_n = cluster_df['reduced_text'][:5]
    for i, row in first_n.iteritems():
        print(f"Row {i}:\n{row}\n")
```
